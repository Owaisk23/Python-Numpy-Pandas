```python
import pandas as pd
from meteostat import Stations, Daily
from datetime import datetime

# Define cities with coordinates (lat, lon)
cities = {
    "Karachi": (24.8607, 67.0011),
    "Lahore": (31.5497, 74.3436),
    "Islamabad": (33.6844, 73.0479),
    "Quetta": (30.1798, 66.9750),
    "Peshawar": (34.0151, 71.5249),
    "Multan": (30.1575, 71.5249),
    "Hyderabad": (25.3960, 68.3578),
    "Skardu": (35.3350, 75.5490)
}

# Date range
start = datetime(2021, 2, 2)
end = datetime(2025, 6, 30)

# Collect data for each city
all_data = []

for city, (lat, lon) in cities.items():
    stations = Stations()
    station = stations.nearby(lat, lon).fetch(1)   # get nearest weather station
    if not station.empty:
        station_id = station.index[0]
        data = Daily(station_id, start, end).fetch()
        if not data.empty:
            data.reset_index(inplace=True)
            data["city"] = city
            data["lat"] = lat
            data["lon"] = lon
            all_data.append(data)

# Combine all city data
df = pd.concat(all_data, ignore_index=True)

# Save to CSV
# df.to_csv("pakistan_cities_weather.csv", index=False)

print(df.head())
print("Dataset shape:", df.shape)
print("Unique cities:", df['city'].unique())
```

```python
import pandas as pd

# Make sure 'time' is datetime
df['time'] = pd.to_datetime(df['time'])

# Loop over each city
for city in df['city'].unique():
    d = df[df['city'] == city].copy()
    d = d.sort_values('time')

    # Create a complete date range from min to max
    full_range = pd.date_range(start=d['time'].min(), end=d['time'].max(), freq='D')

    # Find missing dates
    missing = full_range.difference(d['time'])

    print(f"\nCity: {city}")
    print(f"First date: {d['time'].min().date()}  Last date: {d['time'].max().date()}")
    print(f"Total expected days: {len(full_range)}")
    print(f"Actual days present: {d['time'].nunique()}")
    print(f"Missing days: {len(missing)}")

    if len(missing) > 0:
        print("Example missing dates:", missing[:10])
```

```python
def label_weather(row):
    if row['prcp'] >= 10:
        return "Heavy Rain"
    elif 0 < row['prcp'] < 10:
        return "Light Rain"
    elif row['tmax'] > 35:
        return "Hot"
    elif row['tmin'] < 5:
        return "Cold"
    elif row['wspd'] > 30:
        return "Stormy"
    else:
        return "Sunny/Cloudy"
df4['label'] = df4.apply(label_weather, axis=1).copy()
```

```python

X = df6.drop(['label_d1', 'label_d7',], axis=1)
y1 = df6['label_d1']
y2 = df6['label_d7']




split_idx = int(len(df6) * 0.8)

```

```python
import pandas as pd
import numpy as np
from scipy import stats

numeric_cols = ['tavg', 'tmin', 'tmax', 'prcp', 'wspd', 'pres']


df6[numeric_cols] = df6[numeric_cols].astype(float)


z_scores = np.abs(stats.zscore(df6[numeric_cols].to_numpy(), nan_policy='omit'))


z_scores = pd.DataFrame(z_scores, columns=numeric_cols, index=df6.index)

df_zscore_cleaned = df6[(z_scores < 3).all(axis=1)]

print("Original shape:", df6.shape)
print("After Z-score removal:", df_zscore_cleaned.shape)



def remove_outliers_iqr(data, cols):
    cleaned = data.copy()
    for col in cols:
        Q1 = cleaned[col].quantile(0.25)
        Q3 = cleaned[col].quantile(0.75)
        IQR = Q3 - Q1
        lower = Q1 - 1.5 * IQR
        upper = Q3 + 1.5 * IQR
        cleaned = cleaned[(cleaned[col] >= lower) & (cleaned[col] <= upper)]
    return cleaned

df_iqr_cleaned = remove_outliers_iqr(df6, numeric_cols)

print("After IQR removal:", df_iqr_cleaned.shape)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()


X_scaled = pd.DataFrame(scaler.fit_transform(X), columns=X.columns)
```

```python
X_train, X_test = X_scaled.iloc[:split_idx], X_scaled.iloc[split_idx:]
y1_train, y1_test = y1.iloc[:split_idx], y1.iloc[split_idx:]
y2_train, y2_test = y2.iloc[:split_idx], y2.iloc[split_idx:]

# from sklearn.model_selection import train_test_split

# X1_train, X1_test, Y1_train, Y1_test = train_test_split(
#     X_scaled, y1, test_size=0.1, random_state=42
# )

# X2_train, X2_test, Y2_train, Y2_test = train_test_split(
#     X_scaled, y2, test_size=0.1, random_state=42
# )
```

```python
from imblearn.over_sampling import SMOTE
from collections import Counter


smote1 = SMOTE(random_state=42)
X1_res, y1_res = smote1.fit_resample(X_train, y1_train)
print("Balanced label_d1:", Counter(y1_res))


smote2 = SMOTE(random_state=42)
X2_res, y2_res = smote2.fit_resample(X_train, y2_train)
print("Balanced label_d7:", Counter(y2_res))
```

```python
from sklearn.ensemble import RandomForestClassifier


model_d1 = RandomForestClassifier(class_weight='balanced', random_state=42)
model_d1.fit(X1_res, y1_res)


model_d7 = RandomForestClassifier(class_weight='balanced', random_state=42)
model_d7.fit(X2_res, y2_res)
```

```python
from sklearn.metrics import classification_report
y1_pred = model_d1.predict(X_test)
print("Classification report for label_d1:")
print(classification_report(y1_test, y1_pred))

y2_pred = model_d7.predict(X_test)

print("Classification report for label_d7:")
print(classification_report(y2_test, y2_pred))
```
